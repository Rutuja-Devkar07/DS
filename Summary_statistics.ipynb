Write a Python program to display some basic statistical details like percentile, mean, standard
deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris- versicolor’ of iris.csv dataset.

```python
df.head()
```

```python
def display_statistics(species_df):
    print(f"Statistics for {species_df['Species'].iloc[0]} - PetalLengthCm:")
    # Calculate and display basic statistics for PetalLengthCm
    petal_length = species_df['PetalLengthCm']
    print("Mean:", petal_length.mean())
    print("Standard Deviation:", petal_length.std())
    print("Percentiles (25th, 50th, 75th):\n", petal_length.quantile([0.25, 0.5, 0.75]))
    print("Mode:", petal_length.mode())
    print("Median:", petal_length.median())
    print("Mean: ", petal_length.mean())
    print("\n")

iris_setosa = df[df['Species'] == 'Iris-setosa']
iris_versicolor = df[df['Species'] == 'Iris-versicolor']
iris_virginica = df[df['Species'] == 'Iris-virginica']
```

```python
display_statistics(iris_setosa)
display_statistics(iris_versicolor)
display_statistics(iris_virginica)
```

SKEWNESS

```python
SepallengthCm = df['SepalLengthCm']
print("Mean: ", np.mean(SepallengthCm))
print("Median: ", np.median(SepallengthCm))
print("Standard deviation",np.std(SepallengthCm))
skewness = 3 * (np.mean(SepallengthCm) - np.median(SepallengthCm)) / np.std(SepallengthCm)
print("Skewness: ", skewness)
```

```python
SepalWidthCm = df['SepalWidthCm']
print("Mean: ", np.mean(SepalWidthCm))
print("Median: ", np.median(SepalWidthCm))
print("Standard deviation",np.std(SepalWidthCm))
skewness = 3 * (np.mean(SepalWidthCm) - np.median(SepalWidthCm)) / np.std(SepalWidthCm)
print("Skewness: ", skewness)
```

```python
sns.distplot(SepalWidthCm)
```

```python
numeric_df = df.select_dtypes(include=np.number)
skewness_values = numeric_df.skew()
print(skewness_values)
sns.distplot(skewness_values)
```
